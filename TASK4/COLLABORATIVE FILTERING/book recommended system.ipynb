```python
ratings.rename(columns = {
    "User-ID" : "user_id",
    "Book-Rating" : "rating"
},inplace = True)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

```python
x = ratings["user_id"].value_counts() >200
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
num_rating.head()
```

```python

num_rating.rename(columns={"rating":"num_of_rating"},inplace=True)
```

```python
num_rating.head()
```

```python
ratings_with_books.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on= "title" )
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>50]
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python

distance, suggestion = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
books_name = book_pivot.index
```

```python
import pickle
import os

# Define the directory for saving artifacts
artifact_dir = 'artifacts'
os.makedirs(artifact_dir, exist_ok=True)

# Save the model and data using pickle
pickle.dump(model, open(os.path.join(artifact_dir, 'model.pkl'), 'wb'))
pickle.dump(books_name, open(os.path.join(artifact_dir, 'books.pkl'), 'wb'))
pickle.dump(final_rating, open(os.path.join(artifact_dir, 'final_rating.pkl'), 'wb'))
pickle.dump(book_pivot, open(os.path.join(artifact_dir, 'book_pivot.pkl'), 'wb'))
```

```python
import numpy as np

def recommend_book(book_name):
    try:
        
        book_id = np.where(book_pivot.index == book_name)[0][0]
        
        
        distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
        
        
        for i in range(1, len(suggestion[0])):  
            print(book_pivot.index[suggestion[0][i]])
    
    except IndexError:
        print("Book not found in the pivot table.")
    except Exception as e:
        print(f"An error occurred: {e}")

```

```python
book_name = 'Hemlock Bay'
recommend_book(book_name)
```
